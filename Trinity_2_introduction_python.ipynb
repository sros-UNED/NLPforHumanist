## Loops, iterations

To build a loop in the code we can use  **while** and `for ... in`.

Iterator:
    
>sentence 1

>sentence 2

A loop is execute while the condition is TRUE


```python
week=('Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday')
day=0;
while day<7:
    if day<5:
        print ("Today is {0}. ¡Let´s Work!".format(week[day]))
    else: 
        print ("Today is {0}. It is weekend ".format(week[day]))
    day=day+1
```

Sentence `break` to stop the loop.

```python
day=0;
while day<7:
    if day<5:
        print ("Today is {0}. ¡Let´s Work!".format(week[day]))
    else: 
        print ("Today is {0}. It is weekend ".format(week[day]))
        break
    day=day+1
```

Loop`for ... in ...` is used to itereated over a a iterable elements `( i.e. collection, list, dictionaries...)`

Between `for` and `in` we need a variable to iterate. After `in` we include the collection we want to go through

```python
for day in week:
    print (day)
```

```python
dictionary= {"Spain": 'Sevilla', 'France': "Paris", 'UK': 'London','Ireland':'Dublin','Italy':'Rome'}
for item in dictionary.values():
    print (item)
    
```

## Functions

A function is a block of code which only runs when it is called.

You can pass data, known as parameters, into a function.

A function can return data as a result.
In Python a function is defined using the `def` keyword:


```python
def ddd(a, b):
    """This function add two numbers  """
       #This is a comment as well
    return a+b


ddd(2,3)


```

```python
help(ddd)
```

 It is possible to alter the order of parameters 

```python
def subtraction(a,b):
    """This function substarct two numbers"""
    return a-b

print (subtraction(5,3))
print (subtraction(b=3,a=5))
```

### Some examples of funtions:

    a) inicializated parameters
    b) unknow number of parameters
    c) more da one return value

```python
#A inicializated parameters
def power(exponent,base=2):
    """ This a power function  base^exponent. By default base=2"""
    return base**exponent

print (power(10))
print (power(3,5))
```

```python
#b) unknow number of parameters
def tsum(a,*others):
    """this functions adds all the parameters"""
    sum=a
    for item in others:
        sum=sum+item
    return sum

print (tsum(2,3))
print (tsum(2,3,4,5,6))
```

```python
# c) more than one return value

from math import pi

def circle(radius):
    """Claculte the area od a circle and the lenght of its circunference """
    diameter=radius*2
    area=2*radius*pi
    length=pi*(radius**2)
    return area,length,diameter,radius

d,a,l,r =circle(13)
print ("The radius of the circle is  {3}, its diameter is {2}, and its area is {0:2.2f}. Finally The length of the circunference is   {1:2.2f}".format(d,a,l,r))
```
